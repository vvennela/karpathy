```python
# A Neural Probabilitistic Language Model (Bengio et al 2003) 
# we take a vocabulary of n, and embed it randomly in an x dimensional vector
# using backprop, we then tune the vectors until we get our desired semantic meaning
## for example,  "the dog walked in the room" and "the cat ran in the bedroom" are semantically and syntactically similar
## normally, if the model encounters "a dog ran into the bedroom" we are out of distibution meaning we will get a bad prediction
## but, if the model generalizes these examples, we will still be able to predict the next word 
### we can transfer knowledge through our embedding - the network can understand that "a" and "the" play similar roles 
### similarly, that dogs and cats are animals and can be used in this way in a sentence
# we still want to maximize log likelihood as our cost function
```

```python
import torch 
import torch.nn.functional as F
import matplotlib.pyplot as plt
import numpy as np 
%matplotlib inline
```

```python
words = open("names.txt", "r").read().splitlines()
words[:8]
```

```python
len(words)
```

```python
chars = sorted(list(set("".join(words))))
stoi = {s: i + 1 for i, s in enumerate(chars)}
stoi["."] = 0
itos = {i : s for s , i in stoi.items()}
print(itos)
```

```python
# build dataset 
block_size = 3 # context length, or how many characters do we want to take to predict the next one
X, Y = [], [] 
for w in words:
    print(w)
    context = [0] * block_size
    for ch in w + ".":
        ix = stoi[ch]
        X.append(context)
        Y.append(ix)
        #print("".join(itos[i] for i in context), "------>", itos[ix])
        context = context[1:] + [ix] # crop and append

X = torch.tensor(X)
Y = torch.tensor(Y)
```

```python
def build_dataset(words):
    block_size = 3
    X, Y = [], []
    for w in words:
        context = block_size * [0]
        for ch in w + ".": 
            ix = stoi[ch]
            X.append(context)
            Y.append(ix)
            context = context[1:] + [ix]
    X = torch.tensor(X)
    Y = torch.tensor(Y) 
    print(X.shape, Y.shape)
    return X, Y

import random 
random.seed(42) 
random.shuffle(words)

n1 = int(0.8 * len(words))
n2 = int(0.9 * len(words))
Xtr, Ytr = build_dataset(words[:n1])
Xdev, Ydev = build_dataset(words[n1:n2])
Xte, Yte = build_dataset(words[n2:])

    
```

```python
# Excersise: improve the model beyond a dev loss of 2.17 by 
# a) optimizing LR, 
# b) changing context size, 
# c) changing params, 
# d) changing batch size
```

```python
def build_set(words): 
    block_size = 5
    X, Y = [], []
    for w in words: 
        context = [0] * block_size
        for ch in w + ".": 
            ix = stoi[ch]
            X.append(context)
            Y.append(ix)
            context = context[1:] + [ix] 

    X = torch.tensor(X)
    Y = torch.tensor(Y) 
        
    return X, Y

import random 
random.seed(42)
random.shuffle(words)

n1 = int(.8 * len(words))
n2 = int(.9 * len(words))

Xtr, Ytr = build_set(words[:n1])
Xdev, Ydev = build_set(words[n1:n2])
Xtest, Ytest = build_set(words[n2:])
```

```python
g = torch.Generator().manual_seed(2147483647)
C = torch.randn((27, 15), generator = g)
W1 = torch.randn((75, 300), generator = g)
b1 = torch.randn((300), generator = g) 
W2 = torch.randn((300, 27), generator = g)
b2 = torch.randn((27), generator = g)
params = [C, W1, b1, W2, b2] 
```

```python
sum(p.nelement() for p in params)
```

```python
for p in params:
    p.requires_grad = True
# loss = 0 
```

```python
for i in range(50000): 
    # constructing the minibatches 
    ix = torch.randint(0, Xtr.shape[0], (32,))

    emb = C[Xtr[ix]]
    h = torch.tanh(emb.view(-1, 75) @ W1 + b1)
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits, Ytr[ix])

    for p in params:
        p.grad = None

    loss.backward()
    lr = 0.00005
    for p in params:
        p.data += -lr * p.grad 
loss
# next optimization is to compute the derivative of the loss rate and then apply the local minima algorithmically 
```
